```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import seaborn as sns
from scipy import stats
sns.set(style="ticks")
from statsmodels.graphics.regressionplots import plot_regress_exog, plot_fit, plot_leverage_resid2, influence_plot
```

**SET 6**

```python
Set6 =pd.read_excel ('Assignment4_linear_regresion_data.xlsx', sheet_name='Set 6')
Set6.head()
```

```python
pd.scatter_matrix(Set6, figsize=(20,10), marker = 'o')
```

```python
cm6 = Set6.corr(method='pearson')
R6 = cm6.loc['y','x']
R6
```

```python
model6 = sm.ols('y~x', data = Set6)
model6 = model6.fit()
model6.params
```

```python
fig = plt.figure(figsize = (20,20))
fig = plot_regress_exog(model6, 'x', fig = fig)
```

```python
fig, ax = plt.subplots(figsize=(15,5))
fig = plot_leverage_resid2(model6, ax=ax)
```

```python
model6.summary()
```

OUTPUT1: **R^2 is 0.913 to one which shows that it is a good fit model and the data shows an increasing trend**

OUTPUT2:   
  - Relationship looks linear
  - Absense of influential points
  - Residual variability is constant
  - Residuals are not auto-correlated
  
OUTPUT3: **The value of y will be -0.3059 when x is 0. For each increment of 1 in x, y will increase by 7.0272**

```python
#Remove Outliers
Set6_M = Set6.drop([42,58,28,5,52])
```

```python
model6_M = sm.ols('y~x',data = Set6_M)
model6_M = model6_M.fit()
model6_M.summary()
```

**By removing 5 outliers, the R^2 increased from 0.913 to 0.935 so the outliers do not have a lot of leverage.**
